```python
import numpy as np
from matplotlib import pyplot as plt

def target_pdf(t):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * t ** 2)

def proposal_pdf(t):
    return 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * (t - 4) ** 2)

N = 10000

samples_Y = np.random.normal(loc=4, scale=1, size=N)

weights = target_pdf(samples_Y) / proposal_pdf(samples_Y)

hist, bins = np.histogram(samples_Y, bins=50, weights=weights, density=True)

fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].plot(bins[:-1], hist, label='Approximation')
axs[0].plot(bins[:-1], target_pdf(bins[:-1]), label='Target PDF', linestyle='--')
axs[0].set_xlabel('t', fontweight='bold')
axs[0].set_ylabel('Probability Density', fontweight='bold')
axs[0].legend()
axs[0].grid(True)
axs[0].set_title('Approximation vs Target PDF', fontweight='bold')

axs[1].scatter(samples_Y, weights, alpha=0.5)
axs[1].set_xlabel('Samples (Y)', fontweight='bold')
axs[1].set_ylabel('Weights', fontweight='bold')
axs[1].set_title('Weighted Samples from Importance Sampling', fontweight='bold')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

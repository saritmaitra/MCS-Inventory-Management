```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
```

```python
x = pd.read_csv('/content/manf.csv')
y = pd.read_csv('/content/retail.csv')
z = pd.read_csv('/content/wholesale.csv')
```

```python
x['Period'] = pd.to_datetime(x['Period'], format='%b-%y')
y['Period'] = pd.to_datetime(y['Period'], format='%b-%y')
z['Period'] = pd.to_datetime(z['Period'], format='%b-%y')

x.set_index('Period', inplace=True)
y.set_index('Period', inplace=True)
z.set_index('Period', inplace=True)

plt.figure(figsize=(10, 6))
x['Value'].plot(label='Manufacture')
y['Value'].plot(label='Retail')
z['Value'].plot(label='Wholesale')

plt.legend()
plt.xlabel('Date')
plt.ylabel('Inventories/Sales')
plt.grid()
plt.show()
```
